# Merging all the datafiles to a single datafile

```python
import pandas as pd
```

```python
files = ["ALQ","BPQ","CDQ","DEMO","DIQ","DR1TOT","DR2TOT","HDL", "HSQ", "INS", "MCQ", "PAQ", "SLQ", "SMQ","SMQRTU","TCHOL","TRIGLY","WHQ"]
```

```python
merged_data = pd.read_csv(f"{files[0]}_data.csv")

for filename in files[1:]:
    data_to_merge = pd.read_csv(f"{filename}_data.csv")

    merged_data = pd.merge(merged_data, data_to_merge, on='SEQN', how='outer')

merged_data.to_csv('merged_data_PART1.csv', index=False)

print(merged_data.head())
```

## Q2

```python
files = ["ALQ","BMX","BPQ","BPXO","CDQ","DEMO","DIQ","HSQ","INQ","KIQ_U", "MCQ", "OCQ", "PAQ", "SLQ", "SMQ", "SMQFAM","SMQRTU","SMQSHS","TCHOL","WHQ"]
```

```python
for file in files:
  file1 = pd.read_sas(f"/content/{file}_J.XPT")
  file2 = pd.read_sas(f"/content/P_{file}.XPT")

  common_columns = file1.columns
  for df in [file2]:
    common_columns = common_columns.intersection(df.columns)

  file1_common = file1[common_columns]
  file2_common = file2[common_columns]


  concatenated_data = pd.concat([file2_common, file1_common], ignore_index=True)

  # Save the concatenated dataset to a new CSV file, if needed
  concatenated_data.to_csv(f'{file}_data.csv', index=False)
```

```python
merged_data = pd.read_csv(f"{files[0]}_data.csv")

for filename in files[1:]:
    data_to_merge = pd.read_csv(f"{filename}_data.csv")

    merged_data = pd.merge(merged_data, data_to_merge, on='SEQN', how='outer')

merged_data.to_csv('merged_data_PART2.csv', index=False)

print(merged_data.head())
```

```python
merged_data.dropna(subset=['MCQ160F'], inplace=True)
```

```python
null_percentage = merged_data.isnull().mean() * 100
columns_to_drop = null_percentage[null_percentage > 30].index
df_cleaned = merged_data.drop(columns=columns_to_drop)
df_cleaned.dropna(subset=['MCQ160F'], inplace=True)
df_cleaned
```
